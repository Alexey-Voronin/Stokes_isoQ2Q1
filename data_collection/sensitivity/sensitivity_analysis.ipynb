```python
import sys
sys.path.append('../../')
from sysmg import *
from sysmg.util.data_analysis import conv_factor, collect_residuals

# plotting
import matplotlib.pyplot      as plt

plt.rc('text.latex', preamble=r"""
                                \usepackage[not1,notextcomp,lcgreekalpha]{stix}
                                \usepackage{amsmath}
                                       """)
plt.rc('xtick', labelsize=12)
plt.rc('ytick', labelsize=12)
plt.rc('axes', labelsize=14) 
plt.rc('font', family='serif', size=12)  
plt.rc('text', usetex=True)

#plt.rc('figure', figsize=[4,4])
```

# Vanka 

```python
rho          = np.load("arrays/vanka_sensitrivity_tau_1.05_dirichlet.npy")
N            = 50
OMEGA_OUTER  = np.linspace(0.02, 1., N)
OMEGA_INNER  = np.linspace(0.02, 1., N)

omega_1_opt  = 0.78
omega_0_opt  = 0.74
```

```python
plt.imshow(rho[:,:], cmap='hot',  vmin=np.min(rho), vmax=min(1, np.max(rho)),
           origin="lower", 
           interpolation='bilinear',
           extent=[min(OMEGA_OUTER),max(OMEGA_OUTER),min(OMEGA_INNER),max(OMEGA_INNER)])

ax = plt.gca()
ax.set_xlabel('$\omega_1$') 
ax.set_ylabel('$\omega_0$') 
plt.colorbar()
plt.scatter( omega_1_opt, omega_0_opt, s=50, c='white', marker='x')

#plt.savefig("vanka_112_sens_dirichlet.pdf",  bbox_inches='tight')
plt.savefig("vanka_112_sens_dirichlet.pdf", bbox_inches='tight')

plt.show()
```
